### Exploratory Data Analysis – E-commerce Profitability Dataset

This notebook contains an initial exploratory analysis of e-commerce transaction data obtained from Kaggle. The dataset includes product-level sales records across various platforms (e.g., Amazon, Flipkart, Myntra), with attributes such as SKU codes, selling rates, gross amounts, product sizes, and transaction dates.

The goal of this EDA is to understand the structure and quality of the dataset, identify missing or inconsistent values, and uncover early insights that can guide further cleaning, feature engineering, and profitability analysis.
